```python
import numpy as np
import pandas as pd
from pandas import Series as se
from pandas import DataFrame as df
```

```python
data = pd.read_excel("transportation-to-work-1.xlsx")
```

```python
frame = df(data)
frame[["ind_id", "mode"]]
```

```python
most_used = frame.groupby("mode_name")["pop_mode"].sum()
most_used.sort_values(ascending=False)
```

```python
public_transport_user = frame.loc[
    frame["mode"] == "PUBLICTR", ["race_eth_name", "mode", "pop_mode", "pop_total"]
]
total_public_transport_user = public_transport_user.groupby("race_eth_name")[
    "pop_mode"
].sum()
total_population_per_race = frame.groupby("race_eth_name")["pop_total"].sum()
percent_population = total_public_transport_user / total_population_per_race * 100
percent_population.sort_values(ascending=False).drop("Total")
```
